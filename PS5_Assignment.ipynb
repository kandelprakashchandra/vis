# WORLD GOVERNANCE INDICATORS (WGI):

As per The World Bank, governance consists "the traditions and institutions by which authority in a country is exercised.  This includes the process by which governments are selected, monitored and replaced; the capacity of the government to effectively formulate and implement sound policies; and the respect of citizens and the state for the institutions that govern economic and social interactions among them."


The World Bank hase been producing world governance reports annually since 1996 to gauge the governance situations in almost all the countires across the globe. It condsiders six composite indicators related to governance which are as follows:
1. Voice and Accountability,
2. Political Stability and Absence of Violence/Terrorism, 
3. Government Effectiveness,
4. Regulatory Quality,
5. Rule of Law, and 
6. Control of Corruption

The data are based on "the views of a large number of enterprise, citizen and expert survey respondents in industrial and developing countries".

Sources:

<a href="https://info.worldbank.org/governance/wgi/"> Worldwide Governance Indicators </a>

<a href="https://info.worldbank.org/governance/wgi/pdf/WGI.pdf"> The Worldwide Governance Indicators: Methodology and Analytical Issues  </a>

#LITERACY & GENDER PARITY BY REGIONS AND INCOME GROUPS

```python
# creating a new dataframe to study literacy rate by gender, income groups and regions
literacy =final_merge.groupby(['Region', 'Income group']).agg({'malelit_%': 'mean', 'femlit_%':'mean' })

# Resetting the index so that the index gets lost which leads to inability to plot graphs
literacy = literacy.reset_index()

# creating  a new variable to examine gender parity in education across regions and income groups
#  Education gender parity is the ratio percantage of female literacy to male literacy

literacy['gender parity'] = literacy['femlit_%'] / literacy['malelit_%'] 

literacy.head(10)
```

##LITERACY & GENDER PARITY BY INCOME GROUPS

```python
# Defining the list of variables to plot
variables = ['malelit_%',	'femlit_%',	'gender parity']

# Defining the list of titles for each bar chart
titles = ['Male Literacy Rates Across Income Groups', 
          'Female Literacy Rates Across Income Groups', 
          'Educational Gender Parity Across Income Groups']

fig, axs = plt.subplots(ncols=len(variables), figsize=(20, 4))

# Creating a loop to plot each variable on a separate subplot
for i, var in enumerate(variables):
    sorted_data = literacy.sort_values(by=[var]) # Sorting the data by the variable in ascending order
    sns.barplot(x='Income group', y=var, data=sorted_data, ax=axs[i])
    axs[i].set_title(titles[i])
    axs[i].set_xlabel('')
    axs[i].set_ylabel('')
    axs[i].tick_params(axis='x', rotation=90) 

# Adjusting the spacing between subplots
plt.tight_layout();
```

**Interpretaton**
It is plausible that higher the literacy rate, the greater is the human capital of a country , which translates to more affluence. The bar charts support this thesis. We can see that as the literacy rate for both male and female goes up as we proceed from lower income groups to high income groups.  In addition, the educational dender parity seems much lower in low income region than in any other region. Gender parity is the highest in the high income countries.
